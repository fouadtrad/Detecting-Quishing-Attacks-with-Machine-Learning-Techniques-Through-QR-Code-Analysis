## Loading QR Code Data and Labels

In this section, we load the QR code dataset and its corresponding labels from .pickle files. The dataset consists of 9,987 QR codes, each stored as a 69x69 numpy array, along with their respective labels.

Note: Make sure you unzip the `QuishingDataset.zip` file before running the code.

```python
import pickle

# Load the QR codes array
with open('qr_codes_29.pickle', 'rb') as f:
    qr_codes = pickle.load(f)

# Load the labels array
with open('qr_codes_29_labels.pickle', 'rb') as f:
    qr_labels = pickle.load(f)

# Check the type and shape of the loaded arrays
print(type(qr_codes), qr_codes.shape)
print(type(qr_labels), qr_labels.shape)
```

## Visualizing Sample QR Codes with Labels

Here, we randomly select 10 QR codes from the dataset and visualize them using a 2x5 grid. The title of each QR code represents its corresponding label. This helps in understanding the dataset and verifying its contents.

```python
import matplotlib.pyplot as plt
import numpy as np

# Select 10 random indices
np.random.seed(42)
sample_indices = np.random.choice(len(qr_codes), 10, replace=False)

# Create a 2x5 subplot
fig, axes = plt.subplots(2, 5, figsize=(12, 6))

for i, ax in enumerate(axes.flat):
    index = sample_indices[i]
    ax.imshow(qr_codes[index], cmap='gray')
    ax.set_title(f'Label: {qr_labels[index]}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
